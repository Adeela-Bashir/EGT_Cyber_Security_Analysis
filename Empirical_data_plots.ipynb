```python
#Number of ransomware attacks vs organisation size
import matplotlib.pyplot as plt
import seaborn as sns
# Load data
file_path = '/content/Cyberattacks.xlsx'
data = pd.read_excel(file_path, sheet_name='dataset')

# Filter data for ransomware-related cases based on 'Summary' column
ransomware_data = data[data['Summary'].str.contains("ransomware", case=False, na=False)]

# Specify the order of organization sizes for the plot
organization_size_order = ['Small', 'Medium', 'Large']

# Apply font settings
plt.rcParams.update({
    "font.family": "serif",  # Use a serif font (like Times New Roman)
    "font.size": 14,         # Increase font size
    "axes.labelsize": 18,    # Label size for x and y axes
    "axes.titlesize": 18,    # Title font size
    "xtick.labelsize": 16,   # X-axis tick labels
    "ytick.labelsize": 16,   # Y-axis tick labels
    "legend.fontsize": 16,   # Legend font size
    "figure.titlesize": 16   # Figure title size
})
# Plot the number of ransomware attacks for each organization size
plt.figure(figsize=(10, 6))
sns.countplot(x='Organisation size', data=ransomware_data, order=organization_size_order)
plt.title('Number of Ransomware Attacks vs. Organisation Size')
plt.xlabel('Organisation Size')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
#plt.rcParams['font.size'] = 12
#plt.rcParams['font.weight'] = 'bold'
plt.tight_layout()
plt.savefig(f"Figure_10.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Number of ransomware attacks vs prevention level

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/content/Cyberattacks.xlsx'
data = pd.read_excel(file_path, sheet_name='dataset')

# Apply font settings
plt.rcParams.update({
    "font.family": "serif",  # Use a serif font (like Times New Roman)
    "font.size": 14,         # Increase font size
    "axes.labelsize": 18,    # Label size for x and y axes
    "axes.titlesize": 18,    # Title font size
    "xtick.labelsize": 16,   # X-axis tick labels
    "ytick.labelsize": 16,   # Y-axis tick labels
    "legend.fontsize": 16,   # Legend font size
    "figure.titlesize": 16   # Figure title size
})

# Filter data for ransomware-related cases based on 'Summary' column
ransomware_data = data[data['Summary'].str.contains("ransomware", case=False, na=False)]

# Specify the order of levels on the x-axis
order_levels = ['Low', 'Medium', 'High']

# Plot the number of ransomware attacks for each level in 'Prevention, Detection and Recovery'
plt.figure(figsize=(10, 6))
sns.countplot(x='Prevention, Detection and Recovery', data=ransomware_data, order=order_levels, color='skyblue')
plt.title('Number of Ransomware Attacks per Prevention Level')
plt.xlabel('Prevention, Detection and Recovery Level')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.savefig(f"Figure_13.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Number of ransomware attacks vs level of digital intensity

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/content/Cyberattacks.xlsx'
data = pd.read_excel(file_path, sheet_name='dataset')

# Apply font settings
plt.rcParams.update({
    "font.family": "serif",  # Use a serif font (like Times New Roman)
    "font.size": 14,         # Increase font size
    "axes.labelsize": 18,    # Label size for x and y axes
    "axes.titlesize": 18,    # Title font size
    "xtick.labelsize": 16,   # X-axis tick labels
    "ytick.labelsize": 16,   # Y-axis tick labels
    "legend.fontsize": 16,   # Legend font size
    "figure.titlesize": 16   # Figure title size
})

# Filter data for ransomware-related cases based on 'Summary' column
ransomware_data = data[data['Summary'].str.contains("ransomware", case=False, na=False)]

# Specify the order of levels for 'Level of digital intensity'
digital_intensity_order = ['Low', 'Low-Medium','Medium-High', 'High']

# Plot the number of ransomware attacks for each level of digital intensity
plt.figure(figsize=(10, 6))
sns.countplot(x='Level of digital intensity', data=ransomware_data, order=digital_intensity_order)
plt.title('Number of Ransomware Attacks vs. Level of Digital Intensity')
plt.xlabel('Level of Digital Intensity')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.savefig(f"Figure_11.png", dpi=300, bbox_inches='tight')
plt.show()
```
